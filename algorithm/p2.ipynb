```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn import preprocessing
```

```python
df1 = pd.read_csv('../data/country_vaccinations.csv')
df1['date'] = pd.to_datetime(df1['date'])
df1 = df1.set_index("date")
```

```python
df1
```

```python
df1 = df1.drop(['iso_code', 'vaccines', 'source_name', 'source_website'], axis = 1)
df1
```

```python
df1 = df1.dropna()
df1
```

```python
# date = df1.date
# df1 = df1.drop(['date'], axis = 1)
```

```python
df1.dtypes
```

```python
df1["country"] = df1["country"].astype('category')
df1.country = df1.country.cat.codes
df1
```

```python
kmeans = KMeans(n_clusters=20, random_state=0).fit(df1)
groups = kmeans.labels_
```

```python
unique, counts = np.unique(kmeans.labels_, return_counts=True)
dict(zip(unique, counts))
```

```python
groups
```

```python
df1.loc[:, 'group'] = groups
```

```python
df1
```

```python
df1.index.month.unique()
```

```python
X_train = df1[:'2021-04'].drop(['total_vaccinations'], axis = 1)
y_train = df1.loc[:'2021-04', 'total_vaccinations']

X_test = df1['2021-05'].drop(['total_vaccinations'], axis = 1)
y_test = df1.loc['2021-05', 'total_vaccinations']
```

```python
lr_reg = LinearRegression().fit(X_train, y_train)
lr_pred = lr_reg.predict(X_test)
```

```python
print('LR SCORE: ', r2_score(y_test, lr_pred))
```
